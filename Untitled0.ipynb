### Эксперимент

```python
SEED = 42
N_RUNS = 1
MATRIX_TYPES = [csc_matrix, csr_matrix, dok_matrix, lil_matrix]
ALGORITHMS = (rpq, rpq_bfs)
START_PERCENTS = [pow(2, i) for i in range(-3, 4)]
```

```python
def make_regex_1(labels):
    l = [labels[i % len(labels)] for i in range(4)]
    return Regex(l[0]).union(Regex(l[1]))


def make_regex_2(labels):
    l = [labels[i % len(labels)] for i in range(4)]
    return ((Regex(l[0]).union(Regex(l[1]))).kleene_star()
            .concatenate(Regex(l[2])))


def make_regex_3(labels):
    l = [labels[i % len(labels)] for i in range(4)]
    return ((Regex(l[2]).union(Regex(l[3])))
            .concatenate(Regex(l[2]).union(Regex(l[3]))).kleene_star()
            .concatenate(Regex(l[0]).kleene_star()))


def make_regex_4(labels):
    l = [labels[i % len(labels)] for i in range(4)]
    return (Regex(l[0])
            .concatenate(Regex(l[1]))
            .concatenate(Regex(l[2]))
            .concatenate((Regex(l[3]).union(Regex(l[0]))).kleene_star())
            )


def make_regexes_for_graph(graph):
    l = get_graph_stats_by_name(graph).sorted_edge_labels
    return {
        "(l1 | l2)": make_regex_1(l),
        "(l1 | l2)* l3": make_regex_2(l),
        "(l3 | l4)+ l1*": make_regex_3(l),
        "l1 l2 l3 (l4 | l1)*": make_regex_4(l)
    }


def make_match_all_regex(graph):
    labels = get_graph_stats(graph).edge_labels
    return reduce(lambda acc, x: Regex(x).union(acc), labels, Regex(labels.pop())).kleene_star()


REGEXES = {
    "(l1 | l2)": make_regex_1,
    "(l1 | l2)* l3": make_regex_2,
    "(l3 | l4)+ l1*": make_regex_3,
    "l1 l2 l3 (l4 | l1)*": make_regex_4
}
REGEXES_LIST = list(REGEXES.keys())
```

Очевидно, что наиболее вычислительно емкой частью обоих алгоритмов являются операции над матрицами, сложность которых (для разреженных матриц) прямо пропорциональна количеству вершин и рёбер в графе. Поэтому, для экономии времени, логично будет замерять время работы не на всех графах, а только на одном крупном и использовать регулярное выражение, принимающее любой переход любое количество раз.

```python
def question_1():
    data = {'mtype': [], 'time': [], 'percent': [], 'alg': []}
    graph = load_graph_by_name('core')
    regex = make_match_all_regex(graph)
    for mtype in MATRIX_TYPES:
        for percent in START_PERCENTS:
            for _rpq in ALGORITHMS:
                start_vertices = cfpq_data.generate_multiple_source_percent(graph, percent, seed=SEED)
                if _rpq == rpq_bfs:
                    args = (graph, regex, start_vertices, None, False, mtype)
                elif _rpq == rpq:
                    args = (graph, regex, start_vertices, None, mtype)

                start = time.time()
                _rpq(*args)
                end = time.time()

                data['alg'].append(_rpq.__name__)
                data['mtype'].append(type_to_name[mtype])
                data['time'].append(end - start)
                data['percent'].append(percent)
    return pd.DataFrame(data)


data_q1 = question_1()
```

```python
def q1_plot(df: pd.DataFrame, title: str):
    g = sns.catplot(
        data=df, x='mtype', y='time', row='alg', col='percent', hue='mtype',
        kind='bar', height=4, aspect=.6,
    )
    g.fig.suptitle(title, fontsize=16, y=1.05)
    g.set_axis_labels('', 'Time, s')
    g.set_titles('start={col_name}%')
    g.despine(left=True)
    g._legend.remove()


q1_plot(data_q1[data_q1.alg == 'rpq'], title='Зависимость времени выполнения от типа используемой матрицы для rpq')
q1_plot(
    data_q1[data_q1.alg == 'rpq_bfs'], title='Зависимость времени выполнения от типа используемой матрицы для rpq_bfs')
```

Далее будет использоваться `lil_matrix` т.к. она показывает лучшие результаты

```python
def question_2(
    graphs: list[str] = GRAPHS,
    start_percents: list[float] = START_PERCENTS,
    rpq_fun=(rpq, rpq_bfs),
):
    data = {'time': [], 'graph_name': [], 'percent': [], 'alg': []}
    for graph_name in graphs:
        graph = load_graph_by_name(graph_name)
        regexes = make_regexes_for_graph(graph_name)
        for percent in start_percents:
            start_vertices = cfpq_data.generate_multiple_source_percent(graph, percent, seed=SEED)
            for re_type, regex in regexes.items():
                for _rpq_fun in rpq_fun:
                    for _ in range(N_RUNS):
                        if _rpq_fun == rpq_bfs:
                            args = (graph, regex, start_vertices, None, False, lil_matrix)
                        elif _rpq_fun == rpq:
                            args = (graph, regex, start_vertices, None, lil_matrix)
                        start = time.time()
                        _rpq_fun(*args)
                        end = time.time()
                        data['graph_name'].append(graph_name)
                        data['percent'].append(percent)
                        data['time'].append(end - start)
                        data['alg'].append(_rpq_fun.__name__)
        print(f'finished for {graph_name}')
    return pd.DataFrame(data)


data_q2 = question_2()
```

```python
def q2_plot(df: pd.DataFrame):
    for gname in data_q2.graph_name.unique():
        _df = df[df.graph_name == gname]
        g = sns.catplot(
            data=_df, x='alg', y='time', col='percent', row='graph_name', hue='alg',
            kind='bar', height=4, aspect=.6,
        )
        g.fig.suptitle('', fontsize=16, y=1.05)
        g.set_axis_labels('', 'Time')
        g.set_titles('{row_name}|start={col_name}%')
        g.despine(left=True)
        g._legend.remove()


q2_plot(data_q2)
```

```python
def question_3(
    graphs: list[str] = GRAPHS,
    start_percents: list[float] = START_PERCENTS,
):
    data = {'time': [], 'graph_name': [], 'percent': [], 'per_state': []}
    for graph_name in graphs:
        graph = load_graph_by_name(graph_name)
        regexes = make_regexes_for_graph(graph_name)
        for percent in start_percents:
            start_vertices = cfpq_data.generate_multiple_source_percent(graph, percent, seed=SEED)
            if start_vertices == set():
                start_vertices = cfpq_data.generate_multiple_source(graph, 1, seed=SEED)
            for re_type, regex in regexes.items():
                for per_state in (True, False):
                    for _ in range(N_RUNS):
                        if not per_state:  # Save time by reusing data from q2
                            continue
                        args = (graph, regex, start_vertices, None, per_state, lil_matrix)
                        start = time.time()
                        rpq_bfs(*args)
                        end = time.time()
                        data['graph_name'].append(graph_name)
                        data['percent'].append(percent)
                        data['time'].append(end - start)
                        data['per_state'].append(per_state)
        print(f'finished for {graph_name}')
    return pd.DataFrame(data)


data_q3 = question_3()
```

```python
d = data_q2[['time', 'percent', 'graph_name']]
d['per_state'] = False
data_q3_plus_q2 = pd.concat((data_q3, d))
```

```python
def q3_plot(df: pd.DataFrame):
    for gname in df.graph_name.unique():
        _df = df[df.graph_name == gname]
        g = sns.catplot(
            data=_df, x='per_state', y='time', col='percent', row='graph_name', hue='per_state',
            kind='bar', height=4, aspect=.6,
        )
        g.fig.suptitle('', fontsize=16, y=1.05)
        g.set_axis_labels('', 'Time')
        g.set_titles('{row_name}|{col_name}%')
        g.despine(left=True)
        g._legend.remove()


q3_plot(data_q3_plus_q2)
```
